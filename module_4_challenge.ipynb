## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["line_subtotal"] = df.unit_price * df.qty
df[["unit_price","qty","line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.


order_weight_price = df.apply(lambda order: order['unit_weight']*7 
                              if order['unit_weight']* order ['qty']>50 
                              else order['unit_weight']*10, axis=1)
df['shipping_price'] = order_weight_price * df ['qty']
df[['unit_price','unit_weight','qty','shipping_price']].head(2)


```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
#PS: used line_price instead of total_price (since output example showed line_price and the following steps ask for line_price not total_price from hereon


df['line_price'] = (df['line_subtotal'] + df['shipping_price']) * 1.0925
df['line_price'] = df['line_price'].round(2)
df[['line_subtotal', 'shipping_price','line_price']].head(3)


```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = df['unit_cost']* df['qty'] + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price


df['line_profit'] = df['line_price'] - df['line_cost']
df ['line_profit'] = df['line_profit'].round(2)
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.


def client_totals(client_id,column):    
    client_df = df.loc[df.client_id == client_id, column]
    total = round(client_df.sum(),2)   
    return total     

for client_id in top_5_client_list:     #top_5_client_list pulled from top part 1 variable assignment
    print(f"{client_id} ${client_totals(client_id,'line_price'):.2f}")


```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.



top_5_client_id = {'client_id': top_5_client_list}
                
columns = ['qty', 'shipping_price', 'line_price', 'line_cost', 'line_profit']

for column in columns:
    
    top_5_client_id[column] = [client_totals(client_id,column) if column != 'qty' else int(client_totals(client_id,column)) for client_id in top_5_client_list ]
    
summarized_df = pd.DataFrame(top_5_client_id).sort_values('line_profit', ascending=False)

summarized_df


```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

format_columns = ['shipping_price', 'line_price', 'line_cost', 'line_profit']

def format(amount):
    return f"${amount/1000000:.2f}M"                          

for col in format_columns:
    summarized_df[col] = summarized_df[col].apply(format)
    
summarized_df.head()


rename = {'client_id':'Client ID','qty':'Units','shipping_price':'Shipping','line_price':'Total Revenue','line_cost':'Total Cost','line_profit':'Total Profit',}
                                            
summarized_df_rename = summarized_df.rename(columns=rename)

summarized_df_rename.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest

summarized_df_sorted = summarized_df_rename.sort_values(by='Total Profit', ascending=False)
summarized_df_sorted.head()
```
